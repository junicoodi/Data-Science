```python
import re

def clean_text(text):
    text = str(text).lower()
    text = re.sub(r"http\S+|www\S+", "", text)  # hapus URL
    text = re.sub(r"[^a-zA-Z\s]", "", text)     # hapus angka & simbol
    text = re.sub(r"\s+", " ", text).strip()    # hapus spasi berlebih
    return text

df['Comments'] = df['Comments'].apply(clean_text)
df
```

```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification
from transformers import AutoTokenizer, AutoModel
import torch
import numpy as np
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from transformers import AutoTokenizer, AutoModelForSequenceClassification
```

```python
import torch
import pandas as pd
from transformers import AutoTokenizer, AutoModelForSequenceClassification

# Copy data
df_bert = df_new.copy()

# --- Sentiment dari Comments pakai model pretrained ---
model_name = "nlptown/bert-base-multilingual-uncased-sentiment"
tokenizer = AutoTokenizer.from_pretrained(model_name)
classifier_model = AutoModelForSequenceClassification.from_pretrained(model_name)

def label_num_comments_sentiment(text):
    inputs = tokenizer(text, return_tensors='pt', truncation=True, padding='max_length', max_length=512)
    with torch.no_grad():
        outputs = classifier_model(**inputs)
    probs = torch.nn.functional.softmax(outputs.logits, dim=1)
    pred_class = torch.argmax(probs, dim=1).item() + 1
    if pred_class <= 2:
        return 0  # Negatif
    elif pred_class == 3:
        return 1  # Netral
    else:
        return 2  # Positif

# Tambah label_num_comments
df_bert['label_num_comments'] = df_bert['Comments'].astype(str).apply(label_num_comments_sentiment)

# Ubah label_num_comments ke teks
def label_num_comments_to_text(label_num_comments):
    return ['Negatif', 'Netral', 'Positif'][label_num_comments]

df_bert['label_text_comments'] = df_bert['label_num_comments'].apply(label_num_comments_to_text)

# --- Sentiment dari Rating ---
# Pastikan Rating jadi numerik
df_bert['Rating'] = pd.to_numeric(df_bert['Rating'], errors='coerce')

def label_text_rating(rating):
    if pd.isna(rating):
        return "Unknown"
    elif rating <= 2:
        return "Negatif"
    elif rating == 3:
        return "Netral"
    else:
        return "Positif"

df_bert['label_text_rating'] = df_bert['Rating'].apply(label_text_rating)

# Pilih kolom untuk output akhir
df_bert = df_bert[[
    "Rating",
    "label_text_rating",
    "Date",
    "Category",
    "Comments",
    "label_num_comments",
    "label_text_comments"
]]

df_bert
```

```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.svm import SVC
import seaborn as sns
import matplotlib.pyplot as plt

# --- Data (ambil label dari comments atau rating, pilih salah satu) ---
# Misalnya pakai label dari comments
df_classify = df_bert.copy()
X_text = df_classify['Comments'].astype(str)
y = df_classify['label_num_comments']   # 0 = Negatif, 1 = Netral, 2 = Positif

# --- TF-IDF vektorisasi ---
tfidf = TfidfVectorizer(max_features=5000, ngram_range=(1,2))
X_tfidf = tfidf.fit_transform(X_text)

# --- Train-test split ---
X_train, X_test, y_train, y_test = train_test_split(
    X_tfidf, y, test_size=0.2, random_state=42, stratify=y
)

# --- SVM classifier ---
svm_clf = SVC(kernel="linear", probability=True, random_state=42)
svm_clf.fit(X_train, y_train)

# --- Prediksi ---
y_pred = svm_clf.predict(X_test)

# --- Evaluasi ---
print(classification_report(y_test, y_pred, target_names=["Negatif", "Netral", "Positif"]))

# Confusion Matrix
conf_mat = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(6,5))
sns.heatmap(conf_mat, annot=True, fmt='d', cmap="Blues",
            xticklabels=["Negatif", "Netral", "Positif"],
            yticklabels=["Negatif", "Netral", "Positif"])
plt.xlabel("Prediksi")
plt.ylabel("Aktual")
plt.title("Confusion Matrix - SVM (TF-IDF Features)")
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer

# Copy data
df_vader = df_new.copy()

# Inisialisasi VADER
analyzer = SentimentIntensityAnalyzer()

def vader_sentiment(text):
    scores = analyzer.polarity_scores(str(text))
    compound = scores['compound']
    if compound <= -0.05:
        return 0  # Negatif
    elif compound < 0.05:
        return 1  # Netral
    else:
        return 2  # Positif

# Tambahkan label_num_comments
df_vader['label_num_comments'] = df_vader['Comments'].astype(str).apply(vader_sentiment)

# Ubah label_num_comments ke teks
def label_num_comments_to_text(label_num_comments):
    return ['Negatif', 'Netral', 'Positif'][label_num_comments]

df_vader['label_text_comments'] = df_vader['label_num_comments'].apply(label_num_comments_to_text)

# --- Sentiment dari Rating ---
df_vader['Rating'] = pd.to_numeric(df_vader['Rating'], errors='coerce')

def label_text_rating(rating):
    if pd.isna(rating):
        return "Unknown"
    elif rating <= 2:
        return "Negatif"
    elif rating == 3:
        return "Netral"
    else:
        return "Positif"

df_vader['label_text_rating'] = df_vader['Rating'].apply(label_text_rating)

# Pilih kolom untuk output akhir
df_vader = df_vader[[
    "Rating",
    "label_text_rating",
    "Date",
    "Category",
    "Comments",
    "label_num_comments",
    "label_text_comments"
]]

df_vader
```

## Uji Chi-Square

```python
import pandas as pd
import scipy.stats as stats
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

contingency = pd.crosstab(df_bert['Category'], df_bert['label_text_comments'])
chi2, p, dof, expected = stats.chi2_contingency(contingency)

print("\nHasil Chi-Square Test (Sentiment vs Category):")
print(f"Chi2 = {chi2:.4f}, df = {dof}, p-value = {p:.4f}")

if p < 0.05:
    print("Kesimpulan: Distribusi sentiment berbeda signifikan antar kategori (ada hubungan).")
else:
    print("Kesimpulan: Distribusi sentiment tidak berbeda signifikan antar kategori (tidak ada hubungan).")


# Crosstab Category vs Sentiment
contingency = pd.crosstab(df_bert['Category'], df_bert['label_text_comments'])

# Expected counts dari chi-square
chi2, p, dof, expected = stats.chi2_contingency(contingency)

# Residuals (observed - expected) / sqrt(expected)
residuals = (contingency - expected) / np.sqrt(expected)

print("Residuals (standarized):")
print(residuals)

# Visualisasi residuals → heatmap
plt.figure(figsize=(10,6))
sns.heatmap(residuals, annot=True, cmap="RdBu_r", center=0)
plt.title("Chi-Square Residuals (Category vs Sentiment)")
plt.show()

```

Ada kategori tertentu yang cenderung lebih banyak dapat review positif/negatif.

* Positif (warna biru): Menunjukkan bahwa ada lebih banyak komentar dengan sentimen tersebut daripada yang diharapkan. Misalnya, residual positif 3.3 untuk "Camera" dan "Negatif" menunjukkan bahwa ada lebih banyak komentar negatif daripada yang diharapkan untuk kategori ini.

* Negatif (warna merah tua): Menunjukkan bahwa ada lebih sedikit komentar dengan sentimen tersebut daripada yang diharapkan. Misalnya, residual negatif -2.3 untuk "Camera" dan "Positif" menunjukkan bahwa ada lebih sedikit komentar positif daripada yang diharapkan untuk kategori ini.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Filter untuk kategori Battery dan Display
df_battery = df_bert[df_bert['Category'] == 'Battery']
df_display = df_bert[df_bert['Category'] == 'Display']

# Menghitung distribusi rating untuk masing-masing kategori
rating_battery = df_battery['Rating'].value_counts().sort_index()
rating_display = df_display['Rating'].value_counts().sort_index()

# Menyiapkan data untuk plot
labels = rating_battery.index.union(rating_display.index)

# Membuat dataframe untuk menampung data distribusi
data = pd.DataFrame({
    'Battery': rating_battery.reindex(labels, fill_value=0),
    'Display': rating_display.reindex(labels, fill_value=0)
})

# Visualisasi distribusi rating
plt.figure(figsize=(10, 6))
bars = data.plot(kind='bar', color=['cyan', 'orange'], ax=plt.gca())

# Menambahkan jumlah di atas batang
for bar in bars.patches:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, int(yval), ha='center', va='bottom')

plt.title('Distribusi Rating untuk Kategori Battery dan Display')
plt.xlabel('Rating')
plt.ylabel('Jumlah Komentar')
plt.xticks(rotation=0)
plt.grid(axis='y')
plt.legend(title='Kategori')
plt.show()
```
